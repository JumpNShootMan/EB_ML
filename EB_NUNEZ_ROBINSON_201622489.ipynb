```
import itertools as it
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

### Task 1.3

(3 points) Implement the algorithm you described on the previous task.

```
import pandas as pd
data = pd.read_csv('chess.dat', sep=' ', header=0, skipinitialspace=True)
data.head()
```

```
def load(file):
  with open(file, "r") as f:
    df = []
    for items in f:
      items = items.strip()
      items = items.split(" ")
      df.append(items)
  return df
```

```
df = load("chess.dat")
```

```
print(df)
```

```
te = TransactionEncoder()
te_ary = te.fit(df).transform(df)
data = pd.DataFrame(te_ary, columns=te.columns_)

print(data)
```

```
frequent_itemsets = apriori(data, min_support=0.6, use_colnames=True)
frequent_itemsets
```

### Needed more RAM :(

```
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```
